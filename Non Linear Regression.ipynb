```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas_profiling
```

```python
x = np.arange(-5.0, 5.0, 0.1)
print("Size of x Data is:",x.size)
x
```

```python
#You can adjust the slope and intercept to verify the changes in the graph
y = 2*(x)+3
y_noise = 1 * np.random.normal(size=x.size)
ydata = y+ y_noise

plt.figure(figsize=(16,8))
plt.plot(x,ydata, 'b*')
plt.plot(x,y, 'r')
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

```python
x = np.arange(-5.0, 5.0, 0.1)

y = 1*(x**3)+1*(x**2)+1*x +3
y_noise = 20 * np.random.normal(size=x.size)
ydata = y+ y_noise

plt.figure(figsize=(16,8))
plt.plot(x,ydata, 'b*')
plt.plot(x,y, 'r')
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

# Quadratic

```python
x = np.arange(-5.0, 5.0, 0.1)

y = np.power(x,2)
y_noise = 2 * np.random.normal(size=x.size)
ydata = y+ y_noise

plt.figure(figsize=(16,8))
plt.plot(x,ydata, 'b*')
plt.plot(x,y, 'r')
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.title("Quadratic or plob")
plt.show()
```

# Exponential

```python
x = np.arange(-5.0, 5.0, 0.1)

y = np.exp(x)
y_noise = 5 * np.random.normal(size=x.size)
ydata = y+ y_noise

plt.figure(figsize=(16,8))
plt.plot(x,ydata, 'b*')
plt.plot(x,y, 'r')
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

# Logarthamic

```python
x = np.arange(-5.0, 5.0, 0.1)

y = np.log(x)
y_noise = 0.2 * np.random.normal(size=x.size)
ydata = y+ y_noise

plt.figure(figsize=(16,8))
plt.plot(x,ydata, 'b*')
plt.plot(x,y, 'r')
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

# Sigmoidal/Logistic

```python
x = np.arange(-5.0, 5.0, 0.1)

y = 1-4/(1+np.power(3,x-2))
y_noise = 0.10 * np.random.normal(size=x.size)
ydata = y+ y_noise

plt.figure(figsize=(16,9))
plt.plot(x,ydata, 'bo')
plt.plot(x,y, 'r')
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```
