```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/reddyprasade/Data-Sets-For-Machine-Learnig-and-Data-Science/master/DataSets/Wholesale%20customers%20data.csv')
data.head()
```

```python
from sklearn.preprocessing import Normalizer
```

```python
normal = Normalizer()
```

```python
normal_data = normal.fit_transform(data)
```

```python
normal_data = pd.DataFrame(normal_data,columns= data.columns) 
```

```python
normal_data.head()
```

```python
# import the class containing the clustering method
from sklearn.cluster import AgglomerativeClustering
```

```python
# Create an instance of the class
agg = AgglomerativeClustering(affinity='euclidean', linkage='ward')
```

Fit the instance on the data and then predict clusters for the new data

```python
agg = agg.fit(normal_data)
agg
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(30,50))
plt.title("Dendrograms")
dend = shc.dendrogram(shc.linkage(normal_data, method='ward'))
```

```python
plt.figure(figsize=(10,17))
plt.title("Dendrograms")
dend = shc.dendrogram(shc.linkage(normal_data, method='ward'))
plt.axhline(y=2, color='r', linestyle='--')
```
