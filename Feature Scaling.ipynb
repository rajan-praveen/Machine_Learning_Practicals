```python
np.random.seed(1)
df=pd.DataFrame({
    'x1':np.random.normal(0,2,10000),
    'x2':np.random.normal(5,3,10000),
    'x3':np.random.normal(-5,5,10000)
})
df.head()
```

```python
df.tail()
```

```python
sns.pairplot(df)
```

```python
import pandas_profiling
#pandas_profiling.ProfileReport(df)
```

```python
sns.distplot(df,color=['r','b','g']);
```

## See the plot let assume 0 as mode this this is equaly distributed both side
## means data is normally distributed
## So we r going for "STANDARDSCALER()"

```python
scaler=preprocessing.StandardScaler()
scaled_df=scaler.fit_transform(df)
scaled_df
```

```python
scaled_df=pd.DataFrame(scaled_df,columns=['x1','x2','x3'])
```

```python
sns.distplot(scaled_df,color=['b','r','c'])
```

## After normalize the data from Standardscalar the plot is looking very well distributed

```python
scaled_df.head()
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(20,10))
ax1.set_title("Before Scaling")
sns.kdeplot(df['x1'],ax=ax1)
sns.kdeplot(df['x2'],ax=ax1)
sns.kdeplot(df['x3'],ax=ax1)

ax2.set_title("After standard Scaling")
sns.kdeplot(scaled_df['x1'],ax=ax2)
sns.kdeplot(scaled_df['x2'],ax=ax2)
sns.kdeplot(scaled_df['x3'],ax=ax2)
plt.show()
```

## Positive skew measn most of value is to the left side of mean value(8.18)

```python
import seaborn as sns
sns.distplot(np.random.chisquare(8,1000))
np.mean(np.random.chisquare(8,1000))
```

## Negative skew measn most of value is to the Right side of mean value(31.87)

```python
sns.distplot(np.random.beta(8,2,1000)*40)
np.mean(np.random.beta(8,2,1000)*40)
```

## no skew measn mean, mode , median all r same or nearer to each other and data is normally distributed

```python
sns.distplot(np.random.normal(50,3,1000))
print(np.mean(np.random.normal(50,3,1000)))
print(np.median(np.random.normal(50,3,1000)))
```
