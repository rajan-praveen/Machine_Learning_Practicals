```python
from sklearn.cluster import DBSCAN
from sklearn import metrics
from sklearn.datasets.samples_generator import make_blobs
from sklearn.preprocessing import StandardScaler

# Generate sample data
centers = [[1,1],[-1,-1],[1,-1]]
X, labels_true = make_blobs(n_samples=750, centers=centers, cluster_std=0.4,
                           random_state=0)
X.shape
```

```python
X[0:5]
```

```python
X = StandardScaler().fit_transform(X)
X
```

```python
X.shape
```

```python
# compute DBSCAN
db = DBSCAN(eps=1, min_samples=100).fit(X)
db
```

```python
db.components_
```

```python
db.eps
```

```python
db.labels_
```

```python
db.core_sample_indices_
```

```python
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask
```

```python
core_samples_mask[db.core_sample_indices_]=True
labels = db.labels_
labels
```

```python
len(set(labels))
```

```python
# Number of clusters in labels, ignoring noise if present
n_clusters_= len(set(labels)) - (1 if -1 in labels else 0)
n_clusters_
```

```python
n_noise_ = list(labels).count(-1)
n_noise_
```

```python
metrics.homogeneity_score(labels_true,labels)
```

```python
print('Estimated number of clusters: %d' %n_clusters_)
print('Estimated number of noise points: %d' %n_noise_)
print('Homogeneity: %0.2f' %metrics.homogeneity_score(labels_true, labels))
print('Completeness: %0.3f' %metrics.completeness_score(labels_true, labels))
print('V-measure: %0.3f' % metrics.v_measure_score(labels_true, labels))
print('Adjusted Rand Index: %0.3f'
     %metrics.adjusted_rand_score(labels_true, labels))
print('Adjusted Mutual Information: %0.3f'
     %metrics.adjusted_mutual_info_score(labels_true, labels, average_method='arithmetic'))
#print('Silhoutte Coefficient: %0.3f'% metrics.Sihoutte_score(X, Labels))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,9))

#Black removed and is used for noise instead.
unique_labels = set(labels)
colors = [plt.cm.rainbow(each) for each in np.linspace(0,1,len(unique_labels))]
for k, col in zip(unique_labels, colors):
    if k == -1:
        #Black used for noise.
        col = [0,0,0,1]
        
    class_member_mask = (labels == k)
    
    xy = X[class_member_mask & core_samples_mask]
    plt.plot(xy[:, 0], xy[:,1], '.', markerfacecolor=tuple(col),
            markeredgecolor='r',markersize=50)
    xy = X[class_member_mask & ~core_samples_mask]
    plt.plot(xy[:,0], xy[:,1], '*', markerfacecolor=tuple(col),
            markeredgecolor='k', markersize=15)

plt.title('Estimated number of clusters: %d' %n_clusters_)
plt.show()
```
